## Stationarity Tests

---


```python
# Define a function to determine the number of differences needed
def determine_diffs(series):
    return ndiffs(series.dropna(), test="adf")


# Check for the number of differences needed for stationarity
columns_to_check = [
    "Dollar Rate",
    "Euro Rate",
    "Pound Rate",
    "GCB Monthly Value Traded(GH¢)",
    "CAL Monthly Value Traded(GH¢)",
    "Access Monthly Value Traded(GH¢)",
    "SCB Monthly Value Traded(GH¢)",
    "SOGEGH Monthly Value Traded(GH¢)",
]

diffs_needed = {col: determine_diffs(data[col]) for col in columns_to_check}

print("Number of Differences Needed for Stationarity:")
for col, diffs in diffs_needed.items():
    print(f"{col}: {diffs} difference(s)")
```

```python
# Apply the necessary number of differences
for col, diffs in diffs_needed.items():
    data[f"{col} Diff"] = data[col]
    for i in range(diffs):
        data[f"{col} Diff"] = data[f"{col} Diff"].diff()

# Remove NaN values after differencing
data_diff = data.dropna()
```

```python
data_diff.head()
```

## Modelling and Interpretation


```python
from statsmodels.tsa.ardl import ARDL


# Define a function to fit and summarize ARDL model
def fit_ardl_model(dependent_var, independent_vars, lags=1):
    model = ARDL(dependent_var, lags=lags, exog=independent_vars)
    fitted_model = model.fit()
    return fitted_model.summary()


# Fit ARDL models for each bank
banks = ["GCB", "CAL", "Access", "SCB", "SOGEGH"]
for bank in banks:
    dependent_var = data_diff[f"{bank} Monthly Value Traded(GH¢) Diff"]
    independent_vars = data_diff[
        ["Dollar Rate Diff", "Euro Rate Diff", "Pound Rate Diff"]
    ]
    print(f"ARDL Model for {bank}:")
    print(fit_ardl_model(dependent_var, independent_vars))
```

**Autoregressive Term**: GCB and CAL banks show significant negative coefficients for the lagged traded values, indicating that past traded values negatively influence current traded values for these banks.

**Exchange Rates**: The changes in Dollar, Euro, and Pound rates do not have a statistically significant immediate impact on the monthly traded values of GCB, CAL, Access, SCB, or SOGEGH.
These results suggest that while past traded values can be a significant predictor for some banks (GCB and CAL), the immediate impact of exchange rate changes on traded values is not significant across all banks considered in this analysis.


```python
# Calculate percentage change in exchange rates
data["Dollar Rate Change"] = data["Dollar Rate"].pct_change()
data["Euro Rate Change"] = data["Euro Rate"].pct_change()
data["Pound Rate Change"] = data["Pound Rate"].pct_change()

# Calculate potential profits/losses for each bank
for bank in banks:
    data[f"{bank} Profit from Dollar"] = (
        data[f"{bank} Monthly Value Traded(GH¢)"] * data["Dollar Rate Change"]
    )
    data[f"{bank} Profit from Euro"] = (
        data[f"{bank} Monthly Value Traded(GH¢)"] * data["Euro Rate Change"]
    )
    data[f"{bank} Profit from Pound"] = (
        data[f"{bank} Monthly Value Traded(GH¢)"] * data["Pound Rate Change"]
    )

# Display sample of calculated profits/losses
data[
    ["GCB Profit from Dollar", "GCB Profit from Euro", "GCB Profit from Pound"]
].dropna().head()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns

# List of banks
banks = ["GCB", "CAL", "Access", "SCB", "SOGEGH"]

# Exchange rates columns
exchange_rates = ["Dollar Rate", "Euro Rate", "Pound Rate"]

# Results storage
results = {}


# Function to process each bank
def process_bank(bank_name):
    # Selecting the relevant columns for the bank
    target_col = f"{bank_name} Monthly Value Traded(GH¢)"
    relevant_cols = exchange_rates + [target_col]
    bank_data = data[relevant_cols].dropna()

    # Splitting features and target
    X = bank_data[exchange_rates]
    y = bank_data[target_col]

    # Pipeline setup
    scaler = MinMaxScaler()
    rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

    pipeline = Pipeline(steps=[("scaler", scaler), ("model", rf_model)])

    # Train the model
    pipeline.fit(X, y)

    # Predictions and evaluation
    y_pred = pipeline.predict(X)
    mse = mean_squared_error(y, y_pred)
    mae = mean_absolute_error(y, y_pred)

    # Feature importances
    feature_importances = pipeline.named_steps["model"].feature_importances_

    return mse, mae, feature_importances


# Process each bank
for bank in banks:
    mse, mae, importances = process_bank(bank)
    results[bank] = {"mse": mse, "mae": mae, "importances": importances}

# Display results
for bank, result in results.items():
    print(f"Bank: {bank}")
    print(f"Mean Squared Error: {result['mse']:.3f}")
    print(f"Mean Absolute Error: {result['mae']:.3f}")
    for i, rate in enumerate(exchange_rates):
        print(f"{rate}: {result['importances'][i]:.3f}")
    print("\n")

# Plotting feature importances for each bank
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))
axes = axes.flatten()

for i, bank in enumerate(banks):
    sns.barplot(x=exchange_rates, y=results[bank]["importances"], ax=axes[i])
    axes[i].set_title(f"Feature Importances for {bank}")
    for container in axes[i].containers:
        axes[i].bar_label(container, fmt="%.3f")
    axes[i].set_ylim(0, 1)

plt.tight_layout()
plt.show()
```

#### Relative Importance:

- The Pound Rate has the highest importance (39.1%), followed by the Dollar Rate (35.4%), and then the Euro Rate (25.4%).
- This suggests that fluctuations in the Pound Rate have the most significant impact on the banks' monthly traded values, followed by the Dollar Rate and the Euro Rate.

**Implications**:

- Banks listed on the Ghana Stock Exchange are more sensitive to changes in the Pound and Dollar exchange rates compared to the Euro rate. This could inform risk management strategies, indicating a need for greater focus on mitigating risks associated with the Pound and Dollar exchange rates.
