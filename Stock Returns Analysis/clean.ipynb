```python
import pandas as pd
```

```python
data = pd.read_excel('MTNGH.xlsx')
data.head()
```

```python
cleaned = data[['Daily Date', 'Closing Price - VWAP (GH¢)']]
```

```python
cleaned
```

```python
# Calculate the simple returns
cleaned['Simple Returns'] = cleaned['Closing Price - VWAP (GH¢)'].pct_change()
```

```python
print(cleaned.columns)
```

```python
# Set date as index 
cleaned.set_index('Daily Date', inplace=True) # inplace=True modifies the dataframe in place 
```

```python
cleaned.to_csv('MTNGH_cleaned.csv')
```
