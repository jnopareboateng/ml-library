```python
import pandas as pd
```

```python
# Load your data
factors = pd.read_csv('../data/CleanedData.csv')  # replace with your actual csv file
policy_rate = pd.read_csv('../data/AveragePolicyRate.csv')  # replace with your actual csv file
```

```python
factors.head()
```

```python
policy_rate.head()
```

```python
data = pd.merge(factors, policy_rate, on='Year', how='inner')
data.to_csv('../data/MergedData.csv', index=False)
```

```python
data.head()
```

```python
# set year compoenent as index
data.set_index('Year', inplace=True)

# convert data to time series
data.index = pd.to_datetime(data.index, format='%Y')

```
